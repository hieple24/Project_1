```python
from pathlib import Path
import json
import pandas as pd
import sqlalchemy as sql
from sqlalchemy import create_engine
from time import gmtime, strftime
import time
```

```python
with open(Path("BNB_tweet_count.json")) as f:
    data = json.load(f)
```

```python
data
```

```python
bnb_tweet_count_df = pd.DataFrame.from_dict(data["data"])
```

```python
bnb_tweet_count_df.head()
```

```python
bnb_tweet_count_df["end"] = pd.to_datetime(bnb_tweet_count_df["end"])
```

```python
bnb_tweet_count_df["start"] = pd.to_datetime(bnb_tweet_count_df["start"])
```

```python
bnb_tweet_count_df.head()
```

```python
bnb_tweet_count_df = bnb_tweet_count_df.rename(columns={"end" : "Timestamp"})
```

```python
bnb_tweet_count_df.head()
```

```python
bnb_tweet_count_df.info()
```

```python
engine = create_engine("sqlite:///bnb_tweet_count_df.db", echo=True)
sqlite_connection = engine.connect()
```

```python
sqlite_table = "bnb_tweet_count_df1"
bnb_tweet_count_df.to_sql(sqlite_table, sqlite_connection, if_exists="fail")
```
